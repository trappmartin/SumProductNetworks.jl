# generate test dataset

```julia
using Gadfly

N = 100

X = randn(2, N * 2)
X[:,N+1:end] += 5

Y = zeros(N * 2)
Y[N+1:end] = 1

plot(x = X[1,:], y = X[2,:], color=Y, Geom.point)
```

# construct SPN

```julia
using Distributions
include("../src/SPN.jl")
```

```julia
root = SPN.SumNode(0)

pnode1 = SPN.ProductNode(1, SPN.ClassNode(1))

SPN.add!(pnode1, SPN.UnivariateNode(fit(Normal, X[1,1:N]), 1))
SPN.add!(pnode1, SPN.UnivariateNode(fit(Normal, X[2,1:N]), 2))

pnode2 = SPN.ProductNode(2, SPN.ClassNode(2))

SPN.add!(pnode2, SPN.UnivariateNode(fit(Normal, X[1,N+1:end]), 1))
SPN.add!(pnode2, SPN.UnivariateNode(fit(Normal, X[2,N+1:end]), 2))

SPN.add!(root, pnode1)
SPN.add!(root, pnode2)
```
